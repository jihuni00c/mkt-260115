```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from webdriver_manager.chrome import ChromeDriverManager
from urllib.request import urlretrieve
import time

# 상수값으로서의 변수 목록
URL = "https://moolix.netlify.app/"

service = Service(ChromeDriverManager().install())
options = Options()

options.add_argument("--disable-dev-shm-usage")
options.add_argument("--window-size=1920,1080")
options.add_argument("--start-maximized")
options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36")
options.add_argument("--lang=ko_KR")
options.add_argument("--no-sandbox")

driver = webdriver.Chrome(service=service, options=options)

driver.get(URL)
time.sleep(3)

# search_input = driver.find_element(By.XPATH, "//input[@grid='영화 검색 (제목)']")
# print("검색창 placeholder", search_input.get_attribute("placeholder"))
# elements = driver.find_elements(By.XPATH, "//section/article[@class='card']//div[@class='title']")
movie_cards = driver.find_elements(By.XPATH, "//article[contains(@class, 'card')]")
movie_cards[0].click()
time.sleep(2)

title = driver.find_element(By.XPATH, "//h2").text
date_and_rating = driver.find_element(By.XPATH, "//div[contains(@class, 'muted')]").text
date_text = date_and_rating.split("·")[0].strip()

info = driver.find_element(By.XPATH, "//div[contains(@class, 'stack')]/div[1]").text

print("영화제목 : ", title)
print("개봉연도 : ", date_text)
print("줄거리 : ", info)

driver.quit()

# [4번 문제] xpath -> 해당 사이트 내 첫번째 영화(하우스메이드) 클릭 후 상세페이지 접속 > 제목, 개봉연도, 줄거리 -> 출력
```
