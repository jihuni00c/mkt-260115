```python
# XPATH : 크롤링 선택자의 원조
 
# html = Hyper Text Markup Language => 웹사이트를 구축.구현 -> 마크업태그
# body, div, span, li, ul, a, p...

# XML = Extensable Markup Languge => 사전에 정의되어있는 html 태그보다 유연하게 사용할 수 있는 마크업 문법
# <system> <spacebar>

# 정규화되어있는 태그, 아이디, 클래스, 선택자, 속성을 활용한 크롤링 방법을 보다 유연하게 찾아서 가져오기 위한 목적으로 탄생된 크롤링 방법 (in selenium)
# 시간이 프르면서, xml -> html // xpath -> id, class. selector, attributes 방법으로 크롤링하는 형태가 많아지면서, 요즘 x

# selenium > 크롤링
# 1) find_element VS find_elements
# 2) By, ID, By.CSS_SELECTOR, BY.TAG_NAME, get_attribute()
# 3) By.XPATH

# XPATH 방식으로 특정 값을 찾아오는 문법
"""
/ : 절대경로 -> /html/body/div/div
// : 특정 문서상에서 해당되는 내용을 모두 다 선택 -> //h1 : 문서상 h1태그로 시작되는 값은 다 찾아와라
//*[@href] : 현재 문서에서 href 속성을 가지고 있는 모든 태그를 찾아와라
//a[@href="https://google.com"] : 현재 문서 내 https://google.com 해당 주소값을 가지고 있는 href 속성을 가지고 있는 모든 a를 찾아와라
(//a) [3] : 현재 문서 내 모든 a태그를 찾아와라, 그런데 3번째 등장하는 a만 찾아와라 (주의, 0x, 1부터 시작된다라는 것 주의)
{//table) [last()] : 현재 문서 내 모든 table 태그를 찾아와라, 그런데 마지막번째만 찾아와라
(//a) [position() < 3] : 현재 문서 내 모든 a 태그를 찾아와라, 그런데 3번째 보다 미만인 a만 찾아와라 (첫, 두번째 a만 찾아와라)
//div[@*] : 문서상에서 div태그를 모두 찾아와라, 그런데 속성이 하나라도 존재하는 div태그를 찾아와라 (속성이 무엇이든 상관없음)
contains(@class, "속성") and contains(@class, "속성") -> 복수의 속성을 가지고 있는 경우에 한함
"""
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import time

# 상수값으로서의 변수 목록
URL = "https://davelee-fun.github.io/"

service = Service(ChromeDriverManager().install())
options = Options()

options.add_argument("--disable-dev-shm-usage")
options.add_argument("--window-size=1920,1080")
options.add_argument("--start-maximized")
options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36")
options.add_argument("--lang=ko_KR")
options.add_argument("--no-sandbox")

driver = webdriver.Chrome(service=service, options=options)
driver.get(URL)

# element = driver.find_element(By.XPATH, "/html/body/div[1]/div[1]/div[1]/h1")
elements = driver.find_element(By.XPATH, "//*[@href]")

for element in elements :
    print(element.text)

driver.quit()
```
