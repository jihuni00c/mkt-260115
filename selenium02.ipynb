```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from webdriver_manager.chrome import ChromeDriverManager
import time

# ÏÉÅÏàòÍ∞íÏúºÎ°úÏÑúÏùò Î≥ÄÏàò Î™©Î°ù
URL = "https://moolix.netlify.app/"

service = Service(ChromeDriverManager().install())
options = Options()

options.add_argument("--disable-dev-shm-usage")
options.add_argument("--window-size=1920,1080")
options.add_argument("--start-maximized")
options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36")
options.add_argument("--lang=ko_KR")
options.add_argument("--no-sandbox")

driver = webdriver.Chrome(service=service, options=options)

driver.get(URL)
time.sleep(3)

target_btn = driver.find_element(By.ID, "btnLogin")

target_btn.click()
time.sleep(3)

id_element = driver.find_element(By.ID, "authEmail")
id_element.clear()
id_element.send_keys("sejinpark@gmail.com")

pw_element = driver.find_element(By.ID, "authPw")
pw_element.clear()
pw_element.send_keys("1234")

login_btn = driver.find_element(By.ID, "authSubmit")
login_btn.click()

try :
    WebDriverWait(driver, 5).until(EC.alert_is_present())
    alert = driver.switch_to.alert
    alert.accept()
except :
    print("Alert None")

print(target_btn.text)

driver.quit()
```
