### selenium 문법
0. requests : (find (단일 태그) / find_all (복수 태그) / select (복수의 선택자 : . #) / select_one (단일 선택자))
1. find_elements : 복수의 값을 찾아온다 -> 리스트 (동일한 요소를 전체 다 찾아서 가져온다)
2. find_element : 단수의 값을 찾아온다 -> 위에서 문서를 읽다가, 최초로 발견한 값을 찾아오고 종료
3. By : .TAG_NAME (태그중심), .ID (아이디중심) -> 문서 1번밖에 사용불가, .CSS_SELECTOR (선택자중심 -> 클래스, 태그, 아이디)
4. get.atttribute() : 특정 속성의 속성값을 찾아오고자 할 때
5. .clear() : 특정 영역 내 값을 삭제 (Form 요소 기준 > input[type=radio, check, text], textarea 외)
6. .send_keys() : 어떤 값 혹은 키보드의 키값을 전달
7. Keys : 키보드의 키를 입력하고자 할 때
8. sleep(N) : N초의 시간동안을 지연시키고자 할 때
9. screenshot(filename) : filename으로 스크린캡쳐 가능
10. set_window_size(가로, 세로) : 윈도우
---
requests : html, css, js -> 사이트 문서를 다 받아놓은 후 해당 문서를 다 읽고서 그 안에서 원하는 요소를 찾아오는 방식
selenium : 현재 브라우저에 출력된 데이터를 참고해서 값을 최대한 찾아오는 방식
